```python
#resize
cv2.resize(img,(900,900))
```

```python
#image blur
cv2.blur(img,(5,5))
cv2_imshow(cv2.blur(img,(5,5)))
```

```python
#blurring the image using filter2d (cnn)
cv2_imshow(img)
```

```python
filter1 = np.array([[0.0625,0.125,0.0625],
                  [0.125,0.25,0.125],
                  [0.0625,0.125,0.0625]])
#3x3 gaussian blur filter
```

```python
f= cv2.filter2D(img,-1,filter1)
 # this fun will that 3 parameters the image, the depth and the gaussan filter
cv2_imshow(f)
```

```python
# we can see the above image got blured
```

```python
filter2 = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
#3x3 gaussian sharpen filter
```

```python
f= cv2.filter2D(img,-1,filter2)
 # this fun will that 3 parameters the image, the depth and the gaussan filter
cv2_imshow(f)
```

```python
img=np.ones([200,200,3])*255


cv2.line(img,(0,20),(180,200),(0,0,255),10) # start point, endppoint, color ,thickness
cv2_imshow(img)
```

```python
img=np.ones([200,200,3])*255
cv2.line(img,(0,0),(200,200),(0,0,255),5) # start point, endppoint, color, thickness
cv2.line(img,(100,0),(100,200),(255,0,0),5) # start point, endppoint, color, thickness
cv2.circle(img,(100,100),75,(0,255,0),2) # center, radis,color, thickness
cv2.circle(img,(50,50),50,(255,0,0),2) # center, radis,color, thickness
cv2.rectangle(img,(18,18),(180,180),(100,200,0),2)
cv2_imshow(img)
```
